```python
 # replacing the "MISSING" categories with the modes of teh categorical data
client["channel_sales"].replace({"MISSING" : client.channel_sales.mode()[0]}, inplace = True)
    
client["origin_up"].replace({"MISSING" : client.origin_up.mode()[0]}, inplace = True)
```

```python
# Creating a date function that extracts the duration of the contract in years.

from datetime import date
column_to_convert = ['date_activ', 'date_end', 'date_modif_prod', 'date_renewal'] 
client[column_to_convert] = client[column_to_convert].apply(pd.to_datetime)
client["contract_dur"] = (client["date_end"] - client["date_activ"]).astype('<m8[Y]')
client["durfa_to_ren"] = (client["date_renewal"] - client["date_activ"]).astype('<m8[Y]')
client["acmod_dur"] = np.where(client["date_modif_prod"] >= client["date_activ"],
                               (client["date_modif_prod"] - client["date_activ"]).astype('<m8[Y]'),
                               (client["date_activ"] - client["date_modif_prod"]).astype('<m8[Y]') * -1)
client.drop(column_to_convert, axis = 1, inplace = True)        
```

### Data Visualizations

### Univariate Visualizations

#### Clients Data

```python
columns = ['churn', 'has_gas', 'channel_sales', 'origin_up', 'nb_prod_act', 'forecast_discount_energy']

# Create subplots with the appropriate number of rows and columns
num_columns = len(columns)
num_rows = (num_columns + 1) // 2  # Adjust the number of rows based on the number of columns
fig, axs = plt.subplots(num_rows, 2, figsize=(12, num_rows * 5))

# Iterate over each column and plot the countplot with percentages
for i, column in enumerate(columns):
    row = i // 2
    col = i % 2
    ax = axs[row, col]
    
    sns.countplot(client[column], ax=ax)
    ax.set_title(column)
    sns.despine(ax=ax)
    
    # Calculate the percentage for each category
    total = len(client[column])
    for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height() / total)
        x = p.get_x() + p.get_width() / 2
        y = p.get_height()
        ax.text(x, y, percentage, ha='center', va='bottom')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

### Merging the Prices' and Clients' Data

```python
# merging the client and price data through the id column which is present in both of them
client_price = pd.merge(client_df, price, on = "id")
```

```python
# getting a brief summary on the number of rows with missing values an the datatypes of columns
client_price.info()
```

```python
# confirming of no pressence of nuull values in the merged data
client_price.isnull().sum()
```

```python
# checking for duplicates in teh data
client_price.duplicated().sum()
```

```python
# knowing the total columns in the merged data
client_price.columns
```

```python
# creating new columns that shows the expected change in the price by using forecasted prices and the current prices
client_price['change_energy_price_offpeak'] = client_price['price_off_peak_var'] - client_price['forecast_price_energy_off_peak']
client_price['change_energy_price_peak'] = client_price['price_peak_var'] - client_price['forecast_price_energy_peak']
client_price['change_power_price_offpeak'] = client_price['price_off_peak_fix'] - client_price['forecast_price_pow_off_peak']
```

```python
# re_arranging the data 
new_arrange = ['id', 'channel_sales', 'contract_dur', 'durfa_to_ren', 'acmod_dur',
       'cons_12m', 'cons_gas_12m', 'cons_last_month', 'forecast_cons_12m',
       'forecast_cons_year', 'forecast_discount_energy',
       'forecast_meter_rent_12m','change_energy_price_offpeak','change_energy_price_peak','change_power_price_offpeak', 'has_gas',
       'imp_cons', 'margin_gross_pow_ele', 'margin_net_pow_ele', 'nb_prod_act',
       'net_margin', 'num_years_antig', 'origin_up', 'pow_max',
       'price_date','price_mid_peak_var', 'price_peak_fix',
       'price_mid_peak_fix', 'churn']

cliprice_new = client_price.reindex(columns = new_arrange)
```
